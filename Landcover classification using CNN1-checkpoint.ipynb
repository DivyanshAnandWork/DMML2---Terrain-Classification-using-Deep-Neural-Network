```python
import numpy as np
```

```python
import pandas as pd
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Conv2D, MaxPool2D, Dense, Dropout, Flatten
```

```python
from keras.optimizers import Adam
```

```python
train_chunk = pd.read_csv('/Users/Dell/Desktop/satellite/X_train_sat6.csv',header=None,chunksize=10000)
X_train = pd.concat(train_chunk)
```

```python
x_train_values=X_train.values
```

```python
x_train=x_train_values.reshape(-1,28,28,4)
x_train.shape
```

```python
plt.imshow(x_train[0])
```

```python
test_chunk = pd.read_csv('/Users/Dell/Desktop/satellite/X_test_sat6.csv',header=None,chunksize=10000)
X_test = pd.concat(test_chunk)
```

```python
x_test_values=X_test.values
```

```python
x_test=x_test_values.reshape(-1,28,28,4)
x_test.shape
```

```python
plt.imshow(x_test[0])
```

```python
Y_train=pd.read_csv('/Users/Dell/Desktop/satellite/y_train_sat6.csv',header=None)
```

```python
y_train_values=Y_train.values
```

```python
y_train_values.shape
```

```python
y_train=np.argmax(y_train_values,axis=1)
y_train
```

```python
Y_test=pd.read_csv('/Users/Dell/Desktop/satellite/y_test_sat6.csv',header=None)
```

```python
y_test_values=Y_test.values
```

```python
y_test_values.shape
```

```python
y_test=np.argmax(y_test_values,axis=1)
y_test
```

```python
cnn = Sequential([
    Conv2D(filters=16,kernel_size=(3,3),activation='relu',input_shape=(28,28,4)),
    Conv2D(filters=32,kernel_size=(3,3),activation='relu'),
    MaxPool2D((2,2)),
    Conv2D(filters=64,kernel_size=(3,3),activation='relu'),
    Conv2D(filters=128,kernel_size=(3,3),activation='relu'),
    MaxPool2D((2,2)),
    Conv2D(filters=256,kernel_size=(3,3),activation='relu'),
    MaxPool2D((2,2)),
    Flatten(),
    Dense(60,activation='relu'),
    Dropout(0.5),
    Dense(6,activation='softmax')
])
```

```python
opt=Adam(lr=0.0001,decay=0.0005)
cnn.compile(optimizer=opt,loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
cnn.fit(x_train,y_train,validation_split=0.2,batch_size=128,epochs=10,shuffle=True)
```

```python
   cnn.evaluate(x_test,y_test)
```

```python
y_pred=cnn.predict(x_test)
```

```python
y_pred.shape 
```

```python
y_pred
```

```python
encoder_to_class=[np.argmax(r) for r in y_pred]
encoder_to_class
```

```python
import sklearn
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print("report:\n",classification_report(y_test,encoder_to_class))
```
